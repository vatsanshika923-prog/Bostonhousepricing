```python
##independent and dependent features
X=df.iloc[:,:-1]#skippping last column medv
y=df.iloc[:,-1]#last column
```

```python
X.head()
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#standardizing the  dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

##Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
##print the coefficient and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#on which parameter model has been trained
regression.get_params()
```

```python
##prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
#plot a scater plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
#Residuals
residuals=y_test-reg_pred
```

```python
#plot this residuals
sns.displot(residuals,kind="kde")
```

```python
##Scatter plot with repect to prediction and residuals
##unifor, distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
#adjusted aqaure and r square
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
#display adjusted r squared
1-(1-score)*(len(y_test)-1)/(len(y_test)- X_test.shape[1]-1)
```
